## Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl as openpy
import numpy as np
```

```python
bk_data = pd.read_excel('./Task 2 dataset.xlsx', sheet_name = 'Visualisations')
```

## Insight 1: Which treatments generate the most revenue?

```python
grouped_data = bk_data.groupby(['SpaTreatmentID'])['TotalCostAfterDiscount'].sum()

sorted_data = grouped_data.sort_values(ascending = False)

x_data = sorted_data.index.tolist() # SpaTreatmentID
y_data = sorted_data.values.tolist() # Revenue

plt.bar(x_data, y_data, edgecolor = 'black')

plt.title('Revenue by Spa Treatment', fontsize = 14)
plt.xlabel('Spa Treatment ID', fontsize = 12)
plt.ylabel('Revenue', fontsize = 12)
plt.xticks(rotation = 0)
plt.show()
```

## Insight 2: Appointment Bookings over the years

```python
bk_data['BookingDate'] = pd.to_datetime(bk_data['BookingDate'], format = '%d-%m-%Y')

bk_data['Month'] = bk_data['BookingDate'].dt.month
```

```python
monthly_bookings = bk_data.groupby('Month').size()

plt.plot(monthly_bookings.index, monthly_bookings.values, color='black')

plt.title('Monthly appointment bookings')
plt.xlabel('Month')
plt.ylabel('Number of bookings')
plt.xticks(range(1,13)) # Ensure the x-axis shows all months (1-12)

plt.show()
```

## Insight 3: Most popular spa category by revenue

```python
grouped_data = bk_data.groupby('SpaCategory')['TotalCostAfterDiscount'].sum()

grouped_data
```

```python
plt.pie(grouped_data.values, labels = grouped_data.index, autopct = '%1.1f%%', startangle = 90, shadow = False)

plt.title('Revenue distribution by Spa Category')
plt.show()

```

## Insight 4: Revenue distribution of customers

```python
grouped_data = bk_data.groupby('CustomerID')['TotalCostAfterDiscount'].sum()
```

```python
plt.hist(grouped_data, bins = 10, edgecolor = 'black', color = 'skyblue')
plt.title('Revenue disribution of customers')
plt.xlabel('Total Revenue (£)')
plt.ylabel('Number of Customers')

plt.show()
```
